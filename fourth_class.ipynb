```
# Negative Indexing

email = ["email 1", "email 2" , "email 3" , "email 4"]
print(email[-1])
print(email[-2])
```

```
# Select a range of elements from a list using slicing
# Start index: position where selection begins (inclusive)
# End index: position where selection stops (exclusive, so use index + 1)

prices = [100, 400, 500, 700, 850, 1250, 1500, 2000]

# Get items from index 2 up to index 4 (excluding index 4)
print(prices[2:4])
```
